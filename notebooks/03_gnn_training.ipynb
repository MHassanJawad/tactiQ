# 03 - Tactical GNN Training Curves

This notebook visualizes training dynamics for Task 3A.2.

It reads `checkpoints/gnn_history.pt` produced by `src/training/train_gnn.py` and plots:
- Train vs validation loss
- Train vs validation accuracy

If curves improve over epochs and validation does not collapse, the model is learning useful tactical patterns.

```python
import os
import torch
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-whitegrid")
```

```python
history_path = "../checkpoints/gnn_history.pt"

if not os.path.exists(history_path):
    raise FileNotFoundError(
        f"Could not find {history_path}. Run training first:\n"
        "python -m src.training.train_gnn"
    )

history = torch.load(history_path, map_location="cpu", weights_only=False)
print("Loaded keys:", list(history.keys()))
print("Epochs:", len(history["train_loss"]))
```

```python
epochs = range(1, len(history["train_loss"]) + 1)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(epochs, history["train_loss"], label="Train Loss", linewidth=2)
axes[0].plot(epochs, history["val_loss"], label="Val Loss", linewidth=2)
axes[0].set_title("GNN Loss Curve")
axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("Cross-Entropy Loss")
axes[0].legend()

axes[1].plot(epochs, history["train_acc"], label="Train Accuracy", linewidth=2)
axes[1].plot(epochs, history["val_acc"], label="Val Accuracy", linewidth=2)
axes[1].set_title("GNN Accuracy Curve")
axes[1].set_xlabel("Epoch")
axes[1].set_ylabel("Accuracy")
axes[1].set_ylim(0.0, 1.0)
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
best_epoch = min(range(len(history["val_loss"])), key=lambda i: history["val_loss"][i])

print(f"Best epoch (by val loss): {best_epoch + 1}")
print(f"  Train Loss: {history['train_loss'][best_epoch]:.4f}")
print(f"  Val Loss  : {history['val_loss'][best_epoch]:.4f}")
print(f"  Train Acc : {history['train_acc'][best_epoch]:.4f}")
print(f"  Val Acc   : {history['val_acc'][best_epoch]:.4f}")
```
