# RT-FSAS: Data Exploration
Checking La Liga 2015/16 dataset for graph construction requirements.

```
import json, os
import pandas as pd
import matplotlib.pyplot as plt

data_dir = r"d:\NUST\6th sem\Machine Learning\project\open-data\data"
```

## 1. Load Match List (La Liga 2015/16)

```
with open(os.path.join(data_dir, "competitions.json"), encoding="utf-8") as f:
    comps = json.load(f)
la_liga_1516 = [c for c in comps if c["competition_id"] == 11 and c["season_id"] == 27]
print(f"La Liga 15/16 Found: {bool(la_liga_1516)}")
```

## 2. Inspect Sample Match and Event Schema

```
matches_file = os.path.join(data_dir, "matches", "11", "27.json")
with open(matches_file, encoding="utf-8") as f:
    matches = json.load(f)
print(f"Total matches in season: {len(matches)}")
sample_match_id = matches[0]["match_id"]

with open(os.path.join(data_dir, "events", f"{sample_match_id}.json"), encoding="utf-8") as f:
    events = json.load(f)
print(f"Events in sample match: {len(events)}")

for e in events:
    if "location" in e:
        print("Sample Event:", e["type"]["name"], e["location"])
        break
```

## 3. Check for Three-Sixty Data (Freeze Frames)

```
ts_file = os.path.join(data_dir, "three-sixty", f"{sample_match_id}.json")
has_ts = os.path.exists(ts_file)
print(f"Three-sixty available: {has_ts}\nNOTE: If False, we must fallback to position estimation for off-ball players.")
```

## 4. Profile Season (Events & Types)

```
types_counter = {}
for m in matches:
    ev_file = os.path.join(data_dir, "events", f"{m['match_id']}.json")
    if os.path.exists(ev_file):
        with open(ev_file, encoding="utf-8") as f:
            evs = json.load(f)
        for e in evs:
            t = e["type"]["name"]
            types_counter[t] = types_counter.get(t, 0) + 1

top_types = sorted(types_counter.items(), key=lambda x: x[1], reverse=True)[:5]
print("Top Event Types:", top_types)
```
